```python
import pandas as pd
import numpy as np

df  =pd.read_csv("preprocessed_train.csv", low_memory=False)
df.dropna();
#df = df.head(30000)
reviews = df.text.values;
#print(len(reviews));

#reviews = np.array2string(reviews)
labels = df.label.values;
review_str = [str(i) for i in reviews]
#print(reviews)
df.drop(0,axis = 0)
#print(type(reviews))
#reviews_tokens = [review.split() for review in review_str];
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(review_str, labels, test_size=0.10, random_state=None)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import SGDClassifier
from sklearn import svm
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from sklearn.naive_bayes import MultinomialNB
```

```python
text_clf = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('clf', MultinomialNB()),
                    ])
text_clf = text_clf.fit(X_train, y_train)
```

```python
predicted = text_clf.predict(X_test)
np.mean(predicted == y_test)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {'vect__ngram_range': [(1, 1), (1, 2)],
                  'tfidf__use_idf': (True, False),
                  'clf__alpha': (1e-2, 1e-3),
                 }
gs_clf = GridSearchCV(text_clf, parameters, n_jobs=-1)
gs_clf = gs_clf.fit(X_train, y_train)
gs_clf.best_score_
gs_clf.best_params_
```
